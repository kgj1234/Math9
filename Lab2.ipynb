## Collatz Sequences

Given a number, a collatz sequence is obtained as follows, if the number is divisible by two, the next number in the sequence is the number divided by two. If the number is odd, the next number in the sequence is three times the number plus one.

$13\rightarrow 40\rightarrow 20\rightarrow 10\rightarrow 5\rightarrow 16\rightarrow 8\rightarrow 4\rightarrow 2\rightarrow 1$
